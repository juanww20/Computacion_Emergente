## Pasos a seguir para entrenar un modelo CSV üö©

- **Paso 1**: importar todas las liberias necesarias
- **Paso 2**: Cargar el csv a una tabla de pandas
- **Paso 3**: Definir el Objetivo (y) y las Caracter√≠sticas (X)
- **Paso 4**: Limpiar y transformar (Convertir todo a n√∫meros, manejar nulos).
- **Paso 5**: Dividir (Separar datos de entrenamiento y prueba).
- **Paso 5.5**: Escalar todas las columnas num√©ricas
- **Paso 6**: Construir y entrenar (Alimentar al modelo Keras).
- **Paso 7**: Evaluar (Ver si aprendi√≥ algo).

## Paso 1: importar todas las liberias necesarias

```python
# Librer√≠as para manipulaci√≥n de datos
import pandas as pd

# Librer√≠as de Scikit-Learn para preprocesamiento, modelado y evaluaci√≥n
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc
from sklearn.ensemble import RandomForestClassifier

# Librer√≠as para visualizaci√≥n
import seaborn as sns
import matplotlib.pyplot as plt
#from xgboost import XGBClassifier
#from tensorflow.keras.models import Sequential
#from tensorflow.keras.layers import Dense, Dropout
```

## Paso 2: Cargar el csv

```python
# Cargar el dataset
ruta1 = '../data/student-mat.csv'
df_student_mat = pd.read_csv(ruta1, sep=',')

df_student_mat.head()
df_student_mat.info()
df_student_mat.describe()

print("Columnas:", df_student_mat.columns)
```

## Paso 3: Definir el Objetivo (y) y las Caracter√≠sticas (X)

```python
# Crear la variable 'y'
# 1 = Consumo Alto (Walc > 2)
# 0 = Consumo Bajo/Moderado (Walc <= 2)
df_student_mat['alto_consumo'] = (df_student_mat['Walc'] > 2).astype(int)

y = df_student_mat['alto_consumo']

# Vistazo r√°pido para ver el balance de clases
print(y.value_counts())

# 'X' es todo menos nuestro objetivo 'alto_consumo' y las columnas originales de alcohol
X = df_student_mat.drop(columns=['alto_consumo', 'Walc', 'Dalc'])
```

## Paso 4: Limpiar y transformar

Tienes que convertir todas las columnas de texto (school, sex, Mjob, Fjob, etc.) en n√∫meros.

1. Identificar categ√≥ricas (binarias vs. multi-clase) y num√©ricas.

2. Codificar (Encoding):

    - Usar LabelEncoder o pd.get_dummies para las binarias (ej. sex, address).

    - Usar pd.get_dummies (One-Hot Encoding) para las multi-clase (ej. Mjob, Fjob, reason).

3. Escalar (Scaling):

    - Usar StandardScaler en todas las columnas num√©ricas (ej. age, absences, G1, G2, G3, failures).

```python
# Identificar columnas categ√≥ricas
columnas_categoricas = X.select_dtypes(include=['object']).columns

# Crear listas para cada tipo de codificaci√≥n
# Binarias (Label Encoding)
columnas_binarias = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']

# Nominales (One-Hot Encoding)
columnas_onehot = [col for col in columnas_categoricas if col not in columnas_binarias]

# Aplicar One-Hot Encoding (Crea nuevas columnas)
X = pd.get_dummies(X, columns=columnas_onehot, drop_first=True)

# Aplicar Label Encoding (Convierte M=1, F=0, etc.)
encoder = LabelEncoder()
for col in columnas_binarias:
    if col in X.columns: # Asegurarse de que la columna existe
        X[col] = encoder.fit_transform(X[col])

print("--- Datos despu√©s de codificar ---")
print(X.info()) # ¬°Ahora todo deber√≠a ser num√©rico!
```

## Paso 5: Divisi√≥n de Datos (Entrenamiento y Prueba)

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,     # 20% para prueba
    random_state=42,   # Para reproducibilidad
    stratify=y         # ¬°Importante para clasificaci√≥n!
)

n_features = X_train.shape[1] # Guardar el n√∫mero de columnas
```

## Paso 5.5: Escalar todas las columnas num√©ricas

```python
# 1. Identificar las columnas num√©ricas que NO son categ√≥ricas
# (Las que ya eran n√∫meros en el CSV original)
columnas_numericas = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 
                      'famrel', 'freetime', 'goout', 'health', 'absences', 
                      'G1', 'G2', 'G3']

# 2. Asegurarse de que solo escalamos las columnas que realmente existen
# (despu√©s de get_dummies, algunas pueden haber cambiado, aunque estas no deber√≠an)
cols_a_escalar = [col for col in columnas_numericas if col in X_train.columns]

# 3. Aplicar el escalado
scaler = StandardScaler()

# 4. Ajustar el scaler S√ìLO con los datos de entrenamiento (X_train)
X_train[cols_a_escalar] = scaler.fit_transform(X_train[cols_a_escalar])

# 5. Aplicar la MISMA transformaci√≥n (sin re-ajustar) a los datos de prueba (X_test)
X_test[cols_a_escalar] = scaler.transform(X_test[cols_a_escalar])

print("\n--- Datos despu√©s de escalar ---")
print(X_train[cols_a_escalar].describe())
```

## Paso 6: Construcci√≥n y Compilaci√≥n del Modelo

```python
print("--- Creando Modelo RandomForestClassifier ---")

# Creamos el modelo RandomForest
# n_estimators = 100 (cu√°ntos "√°rboles" construir)
# random_state = 42 (para reproducibilidad)
model_rf = RandomForestClassifier(
    n_estimators=100,
    random_state=120
)

print("Modelo RandomForest creado.")
```

## Paso 7: Entrenamiento y Evaluaci√≥n

```python
# 1. Entrenamiento (API de Scikit-learn)
model_rf.fit(X_train, y_train)

print("--- Evaluaci√≥n del Modelo RandomForest ---")

# 2. Obtener Predicciones
y_pred_clases = model_rf.predict(X_test)
y_pred_probas = model_rf.predict_proba(X_test)[:, 1]  # Probabilidades para la clase positiva

# 3. Evaluaci√≥n
accuracy = accuracy_score(y_test, y_pred_clases)
print(f"Precisi√≥n en prueba: {accuracy:.4f}")

print("\n--- Reporte de Clasificaci√≥n (RandomForest) ---")
print(classification_report(y_test, y_pred_clases, target_names=['Bajo Consumo', 'Alto Consumo']))


print("\n--- Matriz de Confusi√≥n (RandomForest) ---")
cm = confusion_matrix(y_test, y_pred_clases)
print(cm)
plt.figure(figsize=(8, 6))
sns.heatmap(
    cm, 
    annot=True,  # Muestra los n√∫meros (41, 6, 16, 16)
    fmt='d',     # Formato de n√∫meros enteros
    cmap='Blues', # Esquema de color
    xticklabels=['Predijo: Bajo', 'Predijo: Alto'],
    yticklabels=['Real: Bajo', 'Real: Alto']
)

plt.title('Matriz de Confusi√≥n - RandomForest', fontsize=16)
plt.ylabel('Valor Real', fontsize=12)
plt.xlabel('Valor Predicho', fontsize=12)
plt.show()

# Calcular la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_probas)

# Calcular el √Årea Bajo la Curva (AUC)
roc_auc = auc(fpr, tpr)

# Graficar
print("\n--- Curva ROC (RandomForest) ---")
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Adivinanza Aleatoria')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos', fontsize=12)
plt.ylabel('Tasa de Verdaderos Positivos', fontsize=12)
plt.title('Curva "Receiver Operating Characteristic"', fontsize=16)
plt.legend(loc="lower right")
plt.show()
```
